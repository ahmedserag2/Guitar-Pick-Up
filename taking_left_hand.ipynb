```python
import pandas as pd
import numpy as np
```

```python
def generate_labels():
    #NOlandmarks = 5(five fingers)*4(tip,dip,pip,etc)+1(wrist)
    coordinate_labels = ['x','y','z']
    hand_labels = ['left','right']
    wrist_label = 'wrist'
    finger_labels = ['thumb','index','middle','ring','pinky']
    order_labels =['1','2','3','4']
    labels = ['id','frameNo','screen_width','screen_height']
    for hand_label in hand_labels:
        #add wrist
        for coordinate_label in coordinate_labels:
            # zero is put for friendship and error handling
            labels.append(f"{hand_label}_{wrist_label}_0{coordinate_label}")

        for finger_label in finger_labels:
            for order_label in order_labels:
                for coordinate_label in coordinate_labels:
                    labels.append(f"{hand_label}_{finger_label}_{order_label}{coordinate_label}")
    labels.append('label')
    return labels
                
```

```python
#low memory is false because of the warning of column 2,3 dtypes
df = pd.read_csv('data_extracted_for_real.csv',names = generate_labels(),low_memory=False)
df.head()
```

```python
def left_hand(df):
    df_left = df.iloc[:,0:67].copy(deep=True)
    df_left['label'] = df['label']
    return df_left

df_left = left_hand(df)
```
