```python
import pandas as pd 
import numpy as np

import matplotlib as plt 
%matplotlib inline
```

```python
header_list = ['id','frame_id',
               'hand0_wrist_x','hand0_wrist_y',
               'hand0_index_x','hand0_index_y','hand0_middle_x','hand0_middle_y',
               'hand0_ring_x','hand0_ring_y','hand0_pinky_x','hand0_pinky_y',
               'hand1_wrist_x','hand1_wrist_y',
               'hand1_index_x','hand1_index_y','hand1_middle_x','hand1_middle_y',
               'hand1_ring_x','hand1_ring_y','hand1_pinky_x','hand1_pinky_y',
               'target'
              ]
df = pd.read_csv('all_frames.csv',names = header_list)
```

```python
df
```

```python
#left hand limit 1100
#if col values more than 1200 hand is right
df['hand0_wrist_x'].replace('None',np.nan).astype('float64')
```

```python
#df = df[df['hand1_pinky_y'] != 'correct']
#df = df[df['hand1_pinky_y'] != 'totally_wrong']
```

```python
df.iloc[: ,2:22] = df.iloc[: ,2:22].replace('None',np.nan).astype('float64')
```

```python
labels0 = {'hand0_wrist_x':'wrist_x','hand0_wrist_y':'wrist_y',
          'hand0_index_x':'index_x','hand0_index_y':'index_y',
          'hand0_middle_x':'middle_x','hand0_middle_y':'middle_y',
          'hand0_ring_x':'ring_x','hand0_ring_y':'ring_y',
          'hand0_pinky_x':'pinky_x','hand0_pinky_y':'pinky_y'}

labels1 = {'hand1_wrist_x':'wrist_x','hand1_wrist_y':'wrist_y',
          'hand1_index_x':'index_x','hand1_index_y':'index_y',
          'hand1_middle_x':'middle_x','hand1_middle_y':'middle_y',
          'hand1_ring_x':'ring_x','hand1_ring_y':'ring_y',
          'hand1_pinky_x':'pinky_x','hand1_pinky_y':'pinky_y'}
```

```python
#left hands

left_hand0 = df[df['hand0_wrist_x'] < 1100].iloc[:,2:12].copy()
left_hand0.rename(columns = labels0,inplace = True)
```

```python
left_hand1 = df[df['hand1_wrist_x'] < 1100].iloc[:,12:22].copy()
left_hand1.rename(columns = labels1,inplace = True)
```

```python
df.query('hand0_wrist_x > 1100').iloc[:,12:22]
```

```python
#df[df['hand1_wrist_x'].isna()][df['hand0_wrist_x'] > 1100]
left_nans = df.query('hand1_wrist_x.isna() and hand0_wrist_x > 1100',engine='python').iloc[:,12:22].copy()
left_nans.rename(columns = labels1,inplace = True)
```

```python
left_hand1
```

```python
frames = [left_hand0,left_hand1,left_nans]
df_clean = pd.concat(frames)
```

```python
df_clean['id'] = df['id'].copy()
df_clean['frame_id'] = df['frame_id'].copy()
df_clean['target'] = df['target'].copy()
```

```python
df_clean = df_clean[['id','frame_id','wrist_x','wrist_y','index_x','index_y','middle_x','middle_y','ring_x','ring_y','pinky_x','pinky_y','target']]
```
